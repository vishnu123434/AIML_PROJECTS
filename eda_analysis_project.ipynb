#  Outlier detection and removal

```python
Q1 = toyotadf.quantile(0.25)
Q3 = toyotadf.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# slice the dataset that is in IQR range and store in a new dataframe
toyotadf2 = toyotadf[~((toyotadf < (Q1 - 1.5 * IQR)) |(toyotadf > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
toyotadf2.shape
```

## Encode categorical features as a one-hot numeric array.

The input to this transformer should be an array-like of integers or strings, denoting the values taken on by categorical (discrete) features. The features are encoded using a one-hot (aka ‘one-of-K’ or ‘dummy’) encoding scheme. This creates a binary column for each category and returns a sparse matrix or dense array (depending on the sparse parameter)

```python
#ENCODING CATEGORICAL DATA using Get_Dummies
onehot_stoyota = pd.get_dummies(toyotadf, columns = ['FuelType'])
print(onehot_stoyota)
```
